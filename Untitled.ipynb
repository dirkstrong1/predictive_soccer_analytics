```python
engine = create_engine('postgresql:///soccer_database')
df_draw_league = pd.read_sql_query(f"""SELECT club, SUM(draws)/SUM(win+loss+draws) AS draw_percent
                         FROM team_stats 
                         WHERE league = 'England Premier League'
                         GROUP BY club
                         ORDER BY draw_percent DESC;""", engine)
```

```python
epl = df_draw_percent_builder('draw_percent', 'England Premier League')
names = ['Spain LIGA BBVA', 'Germany 1. Bundesliga', 'Italy Serie A']
for name in names:
    df_leagues = []
    name = df_draw_percent_builder('draw_percent', name)
    df_leagues.append(name)
    
    for league, name in zip(df_leagues,names):
        ttests = []
        ttest_results = league_ttest(epl, league, 'draw_percent', 0.05)
        ttests.append(ttest_results)
        print(f"{name}", ttests)
```
