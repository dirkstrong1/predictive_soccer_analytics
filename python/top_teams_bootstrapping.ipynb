```python
import pandas as pd
from sqlalchemy import create_engine
from top_team_stats import df_top_builder, area_plotter
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
league_names = ['Netherlands Eredivisie', 'Scotland Premier League', 'Poland Ekstraklasa', 
                'Portugal Liga ZON Sagres','England Premier League', 'Germany 1. Bundesliga', 
                'Italy Serie A', 'Switzerland Super League','Spain LIGA BBVA', 
                'France Ligue 1', 'Belgium Jupiler League']

league_top_dataframes = {}

for name in league_names:
    league_top_dataframes[name] = df_top_builder(name)
```

```python
y_axis = ['win', 'loss', 'draws']
area_plotter(league_top_dataframes, y_axis, league_names)
```

```python
league_top_dataframes['Belgium Jupiler League']
```

```python
win_percents = {}

for name in league_names:
    game_total = (league_top_dataframes[name][['win', 'loss', 'draws']].sum()).sum()
    win_total = (league_top_dataframes[name][['win']].sum()).sum()
    win_percent = round((win_total/game_total)*100, 3)
    win_percents[name] = win_percent
```

```python
#turning a dictionary of leagues and win_percent to pandas dataframe

df_winpercent = pd.DataFrame(list(win_percents.items()), columns = ['Leagues', 'win_percent'])
```

```python
# bar plot of leagues and their average win percentages over 8 seasons
df_winpercent.plot.bar('Leagues', 'win_percent', color= 'red')
plt.legend(loc='upper right')
plt.ylabel('win_percent')
plt.ylim(0,92);
```

```python
#selecting teams minus EPL
condition_minus_epl = [lis for lis in range(0, 11) if lis !=4] 

#pandas win_percent dataframe with all teams except EPL
df_winpercent_minus_epl = df_winpercent.iloc[condition_minus_epl] 

#pandas win_percent dataframe for EPL
df_winpercent_epl = df_winpercent.iloc[[4]]
```

```python
n_bootstraps = 10000

def bootstrap_proportion(data, n_bootstraps=n_bootstraps):
    """Return array of sample means from bootstrapping n_samples times."""
    output = np.zeros(n_bootstraps)
    n = len(data)
    for i in range(n_bootstraps):
        bootstrap_sample = np.random.choice(data, 
                                            size=n,
                                            replace=True)
        output[i] = (bootstrap_sample.sum() / n)
    return output
```

```python
data_winpercent_leagues=df_winpercent_minus_epl['win_percent']
```

```python
data_winpercent_bootstrap = bootstrap_proportion(data_winpercent_leagues, 1000)
```

```python
sns.distplot(data_winpercent_bootstrap)
plt.xlabel('win_percent')
plt.ylabel('density')
plt.title('bootstrap league champions win percent')
plt.axvline(67.982, color='red', label='EPL') #EPL champion win percent in 8 seasons is 67.9
plt.legend();
```

```python
z_epl_winpercent = (67.982-data_winpercent_bootstrap.mean())/data_winpercent_bootstrap.std()
round(z_epl_winpercent,3)
```

* **EPL is -0.583 away from the mean of win_percent other champions in European leagues**
