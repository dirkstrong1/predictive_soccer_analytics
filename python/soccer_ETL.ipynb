```python
conn = sqlite3.connect('soccer_database.sqlite')
```

```python
df = pd.read_sql_query('''SELECT Date, Season, team_long_name AS home_team, 
                        home_team_goal, away_team_goal, 
                        home_team_api_id, away_team_api_id
                        FROM Match
                        JOIN Team
                        ON Match.home_team_api_id = Team.team_api_id
                        WHERE league_id=1729;
                        ''', conn)
        
```

```python
df.head()
```

```python
df = pd.read_sql_query('''SELECT Date, team_long_name AS away_team, 
                        home_team_goal, away_team_goal, 
                        home_team_api_id, away_team_api_id
                        FROM Match
                        JOIN Team
                        ON Match.away_team_api_id = Team.team_api_id
                        WHERE league_id = 1729;
                        ''', conn)
```

```python
engine = create_engine('postgresql:///soccer_database')
```

```python
df.to_sql("player", engine, if_exists='replace')
```
